```python
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
#SHAREFILE CHILE
gdf = gpd.read_file(r'data_sharefiles\Regiones\Regional.shp')
print(gdf.head(15))
```

```python
chile_regiones = gdf[gdf['Region'].notnull()]
print(chile_regiones.head())
```

```python
chile_regiones.plot(figsize=(10, 10), edgecolor='black')
plt.title('Regiones de Chile')
plt.show()
```

```python
print(gdf.columns)
```

```python
print(gdf.head())
```

```python
# Comprobar el tipo de geometría
print(gdf.geom_type)

# Ver una muestra de las geometrías
print(gdf.geometry.head())
```

```python
#SHAREFILE ARGENTINA

gdf_provincia = gpd.read_file(r'data_sharefiles\provincia\provincia.shp') 

print(gdf_provincia.head())
```

```python
print(gdf_provincia.columns)
```

```python
provincias_argentina = gdf_provincia[gdf_provincia['nam'].notnull()]
print(provincias_argentina.head())
```

```python
provincias_argentina.plot(figsize=(12, 12), edgecolor='black')
plt.title('Provincias de Argentina')
plt.show()
```

```python
# Comprobar el tipo de geometría
print(gdf_provincia.geom_type)
print(gdf_provincia.geometry.head())
```
